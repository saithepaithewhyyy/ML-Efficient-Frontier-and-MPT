```python
# mean_list = np.empty([])
# var_list = np.empty([])

mean_list = []
var_list = []
```

```python
mean_list = np.append(mean_list, [1, 2, 3])
```

```python
test_weights = pd.DataFrame()

kappa = 9
for column in test_proba.columns:
  test_weights[column] = test_proba[column].apply(lambda x: 0 if x < kappa/10 else 1)



row_sum = test_weights.sum(axis=1)

test_weights = test_weights.div(row_sum,axis=0)
test_weights = test_weights.fillna(0)

portfolio = test_return.mul(test_weights)

tot_return = portfolio.sum(axis=1)
tot_return.reset_index(drop=True, inplace=True)

window = 250
for i in range(0, len(tot_return), window):
    window_start = i
    window_end = min(i + window, len(tot_return))
    window_values = tot_return[window_start:window_end]
    mean = window_values.mean()
    variance = window_values.var()
    # mean_list = np.append(mean_list, mean)
    # var_list = np.append(var_list, variance)
    mean_list.append(mean)
    var_list.append(variance)

# sharpe_ratio = (mean_list - 0.015)/var_list
```

```python
len(var_list)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_features = PolynomialFeatures(degree=3)

X = np.array(var_list).reshape(-1, 1)
y = np.array(mean_list)

X_poly = poly_features.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

# Generate predictions
x_plot = np.linspace(min(X), max(X), 100)
y_plot = model.predict(poly_features.fit_transform(x_plot.reshape(-1, 1)))

# Plot the data points and the best fit curve
plt.figure(figsize=(8, 6))
plt.scatter(var_list, mean_list, color='blue', marker='o', label='Data Points')
plt.plot(x_plot, y_plot, color='red', label='Degree 2 Best Fit Curve')

plt.ylabel('Mean')
plt.xlabel('Variance')
plt.legend()

plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.show()
```
